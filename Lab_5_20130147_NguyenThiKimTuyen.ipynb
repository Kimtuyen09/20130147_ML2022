````python
# Import scikit-learn dataset library
from sklearn import datasets
#Import svm model
from sklearn import svm
# Load dataset
cancer = datasets.load_breast_cancer()
````

````python
# code
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.2, random_state=42)

# #Create a svm Classifier
clf = svm.SVC(kernel='linear')

#Train the model using the training sets
clf.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)

# Evaluate the performance of the classifier using accuracy, precision, recall, and f1-score
print('Accuracy: {:.2f}'.format(accuracy_score(y_test, y_pred)))
print('Precision: {:.2f}'.format(precision_score(y_test, y_pred)))
print('Recall: {:.2f}'.format(recall_score(y_test, y_pred)))
print('F1-score: {:.2f}'.format(f1_score(y_test, y_pred)))
````

````python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.2, random_state=42)

# Create classifiers for Logistic Regression, Decision Tree, and kNN
logreg = LogisticRegression()
dtree = DecisionTreeClassifier()
knn = KNeighborsClassifier()

# Train the classifiers on the training data
logreg.fit(X_train, y_train)
dtree.fit(X_train, y_train)
knn.fit(X_train, y_train)

# Make predictions on the testing data for each classifier
y_pred_logreg = logreg.predict(X_test)
y_pred_dtree = dtree.predict(X_test)
y_pred_knn = knn.predict(X_test)

# Evaluate the performance of each classifier using accuracy, precision, recall, and f1-score
print('Logistic Regression:')
print('Accuracy: {:.2f}'.format(accuracy_score(y_test, y_pred_logreg)))
print('Precision: {:.2f}'.format(precision_score(y_test, y_pred_logreg)))
print('Recall: {:.2f}'.format(recall_score(y_test, y_pred_logreg)))
print('F1-score: {:.2f}'.format(f1_score(y_test, y_pred_logreg)))
print('\n')

print('Decision Tree:')
print('Accuracy: {:.2f}'.format(accuracy_score(y_test, y_pred_dtree)))
print('Precision: {:.2f}'.format(precision_score(y_test, y_pred_dtree)))
print('Recall: {:.2f}'.format(recall_score(y_test, y_pred_dtree)))
print('F1-score: {:.2f}'.format(f1_score(y_test, y_pred_dtree)))
print('\n')

print('kNN:')
print('Accuracy: {:.2f}'.format(accuracy_score(y_test, y_pred_knn)))
print('Precision: {:.2f}'.format(precision_score(y_test, y_pred_knn)))
print('Recall: {:.2f}'.format(recall_score(y_test, y_pred_knn)))
print('F1-score: {:.2f}'.format(f1_score(y_test, y_pred_knn)))
print('\n')

# Evaluate the performance of the SVM classifier using accuracy, precision, recall, and f1-score
svm = SVC(kernel='linear')
svm.fit(X_train, y_train)
y_pred_svm = svm.predict(X_test)
print('SVM:')
print('Accuracy: {:.2f}'.format(accuracy_score(y_test, y_pred_svm)))
print('Precision: {:.2f}'.format(precision_score(y_test, y_pred_svm)))
print('Recall: {:.2f}'.format(recall_score(y_test, y_pred_svm)))
print('F1-score: {:.2f}'.format(f1_score(y_test, y_pred_svm)))
````
