#2. Load dataset

#5. Group dataset by ***Gender*** and ***Age***

```python
# code
print (dataset.groupby(['Genre','Age']).groups)
```

#6. Use scatter plot to represent the relationship between ***Age*** and ***Annual Income***
Rememer adding titles, xlabel, ylabel, ... to the plot

```python
# code
ages = dataset["Age"]
annualIncome = dataset["Annual Income (k$)"]
plt.scatter(ages, annualIncome, label= "starts", color="blue", marker="*", s=20)
plt.xlabel('ages') 
plt.ylabel('annualIncome')  
plt.title('My scatter plot!') 
plt.legend() 
plt.show()
```

#7. Use histogram plot to represent the distribution of ***Age***

```python
#code
ages = dataset["Age"]
plt.hist(ages, bins = 10, color = 'blue', alpha = 0.5)
plt.show()
```
