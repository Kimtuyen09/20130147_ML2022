````python
from sklearn.datasets import load_iris
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from prettytable import PrettyTable



````

````python
#code
iris = load_iris()

param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf','linear']}
svm = SVC()
grid_search = GridSearchCV(svm, param_grid, cv = 5)
grid_search.fit(iris.data, iris.target)      
print("Best parameters: " , grid_search.best_params_)        
````

*  1.2. Apply **GridSearchCV** for **kNN** to find the best hyperparameters using the following param_grid.

```
grid_params = { 'n_neighbors' : [5,7,9,11,13,15],
               'weights' : ['uniform','distance'],
               'metric' : ['minkowski','euclidean','manhattan']}
```
where

    *  **n_neighbors**: Decide the best k based on the values we have computed earlier.
    *  **weights**: Check whether adding weights to the data points is beneficial to the model or not. 'uniform' assigns no weight, while 'distance' weighs points by the inverse of their distances meaning nearer points will have more weight than the farther points.
    *  **metric**: The distance metric to be used will calculating the similarity.


````python
#code
iris = load_iris()
grid_params = { 'n_neighbors' : [5,7,9,11,13,15],
               'weights' : ['uniform','distance'],
               'metric' : ['minkowski','euclidean','manhattan']}
knn = KNeighborsClassifier()
grid_search = GridSearchCV(knn, grid_params,cv =5)
grid_search.fit(iris.data, iris.target)
print("Best parameters: ", grid_search.best_params_)
````

*   1.4 Compare the best obtained results from 1.1 to 1.3 (use PrettyTable to dispaly the results)

````python
 # Load the iris dataset
iris = load_iris()

# Define the SVM estimator and parameter grid
svm = SVC()
svm_params = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf','linear']}

# Define the kNN estimator and parameter grid
knn = KNeighborsClassifier()
knn_params = {'n_neighbors': [5, 7, 9, 11, 13, 15],
              'weights': ['uniform', 'distance'],
              'metric': ['minkowski', 'euclidean', 'manhattan']}

# Define the Random Forest estimator and parameter grid
rf = RandomForestClassifier()
rf_params = {'n_estimators': [10, 50, 100, 200],
             'max_depth': [None, 5, 10, 20, 30]}

# Create the GridSearchCV objects
grid_svm = GridSearchCV(svm, svm_params, cv=5)
grid_knn = GridSearchCV(knn, knn_params, cv=5)
grid_rf = GridSearchCV(rf, rf_params, cv=5)

# Fit the GridSearchCV objects to the data
grid_svm.fit(iris.data, iris.target)
grid_knn.fit(iris.data, iris.target)
grid_rf.fit(iris.data, iris.target)

# Create the table
table = PrettyTable(['Classifier', 'Best Parameters', 'Best Score'])

# Add rows to the table
svm_params = grid_svm.best_params_
svm_score = grid_svm.best_score_
table.add_row(['SVM', svm_params, svm_score])

knn_params = grid_knn.best_params_
knn_score = grid_knn.best_score_
table.add_row(['kNN', knn_params, knn_score])

rf_params = grid_rf.best_params_
rf_score = grid_rf.best_score_
table.add_row(['Random Forest', rf_params, rf_score])

# Print the table
print(table)
````
